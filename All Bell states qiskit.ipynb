```python
from qiskit import*
import matplotlib.pyplot as plt
from qiskit.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor
```

```python
#Bell state for two qubit
bell = QuantumCircuit(2,2)
bell.h(0)
bell.cx(0,1)

bell.measure([0,1], [0,1])
```

```python
bell.draw()
```

```python
#run quantum circuit
backend = BasicAer.get_backend('qasm_simulator')  #will run on
circ = bell.compose(bell)
result = backend.run(transpile(circ, backend),shots=2000).result()
counts = result.get_counts(circ)
print(counts)
```

```python
plot_histogram(counts)
```
